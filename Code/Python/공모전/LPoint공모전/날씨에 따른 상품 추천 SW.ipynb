```python
#거래 정보 데이터
data2 = pd.read_csv("./data/data_transaction.csv", encoding='utf-8')
trans_df = pd.DataFrame(data2) 
trans_df.sort_values(by=['de_dt'], axis=0)
trans_df.rename(columns = {"de_dt" : "date"}, inplace = True)
trans_df.head(10)
```

```python
# 날씨 데이터와 거래정보 데이터 병합 (기준: 날짜)
join_df = pd.merge(left=trans_df, right=weather_df, on="date", how="left")
join_df = join_df.sort_values(by=["date", "de_tm", "clnt_id"])
join_df = join_df[join_df["pd_c"] != "unknown"]
join_df.head(10)
```

```python
# 날짜 list & dictionary
date_len = len(join_df["date"].unique())
date_list = [elem for elem in join_df["date"].unique()]
date_list.sort()
date_dict = dict()
for idx, elem in enumerate(date_list): 
    date_dict[elem] = idx
date_len, date_list
```

```python
# 상품코드 dictionary (key: 상품 코드, value: 상품 판매량 리스트, 인덱스 = 날짜 => 0번째 인덱스: 20190701 )
pd_dict = dict() 
for i in range(len(join_df)): 
    row = join_df.iloc[i] 
    pd_c, date = row["pd_c"], row["date"]
    if pd_c not in pd_dict: pd_dict[pd_c] = [0]*date_len
    pd_dict[pd_c][date_dict[date]] += 1
pd_dict
```

```python
#점이연 상관관계 point biserial (연속형 변수 : 이분형 변수 상관계수)
from scipy.stats import pointbiserialr

#비와 판매된 상품 상관 관계 분석
rain_info = weather_df["weather_비"]
result_dict = dict()
for pd_c in pd_dict: 
    current = pd_dict[pd_c] 
    pb = pointbiserialr(rain_info, current)
    result_dict[pd_c] = pb
```

```python
import math 
result = list() 
for key, val in result_dict.items():
    if not math.isnan(val[0]):
        if val[1] <= 0.05: result.append((key, *val))
result.sort(key=lambda x: x[1], reverse=True)  # x[0]: 상관계수, x[1]: p-value
result
```

```python
# 미세먼지와 판매된 상품 상관관계 Pearson 사용 (연속형 변수 : 연속형 변수)
from scipy.stats import pearsonr
dust_info = weather_df["ave_dust"]
```

```python
nan_idx = list() # Nan 값 들어있는지 확인
for idx, elem in enumerate(dust_info):
    if math.isnan(elem): 
        nan_idx.append(idx) 
dust_info = dust_info.dropna()
nan_idx, len(dust_info) #[] -> Nan값 없음
```

```python
result_dict = dict()
for pd_c in pd_dict: 
    current = pd_dict[pd_c] 
    ps = pearsonr(dust_info, current)
    result_dict[pd_c] = ps
result_dict
result = list() 
for key, val in result_dict.items():
    if val[1] <= 0.05: result.append((key, *val))
result.sort(key=lambda x: x[1], reverse=True)

dust_rcm=[]
for elem in result: 
    pd_c = elem[0] 
    name = product_df[product_df["pd_c"] == int(pd_c)]
    name = name["clac_nm3"].values[0]
    print(elem, name)    # 미세먼지 피어슨 상관계수 r, p-value
    if elem[1] > 0.3 and elem[2] < 0.05:
        dust_rcm.append(name)
```

```python
#흐림 판매된 상품 상관 관계 분석 (Point Biserial correlation)
cloudy_info = weather_df["weather_흐림"]
result_dict_cld = dict()
for pd_c in pd_dict: 
    current = pd_dict[pd_c] 
    pb_cld = pointbiserialr(cloudy_info, current)
    result_dict_cld[pd_c] = pb_cld
    
result_cld = list() 
for key, val in result_dict_cld.items():
    if not math.isnan(val[0]):
        if val[1] <= 0.05: result_cld.append((key, *val))
result_cld.sort(key=lambda x: x[1], reverse=True)

cloudy_rcm=[]
# 흐린 날과 판매된 상품의 상관 계수, p-value
for elem in result_cld: 
    pd_c = elem[0] 
    name = product_df[product_df["pd_c"] == int(pd_c)]
    name = name["clac_nm3"].values[0]
    print(elem, name)
    if elem[1] > 0.3 and elem[2] < 0.05:
        cloudy_rcm.append(name)
```

```python
#맑음 판매된 상품 상관 관계 분석 (Point Biserial correlation)
sunny_info = weather_df["weather_맑음"]
result_dict_sun = dict()
for pd_c in pd_dict: 
    current = pd_dict[pd_c] 
    pb_sun = pointbiserialr(sunny_info, current)
    result_dict_sun[pd_c] = pb_sun
    
result_sun = list() 
for key, val in result_dict_sun.items():
    if not math.isnan(val[0]):
        if val[1] <= 0.05: result_sun.append((key, *val))
result_sun.sort(key=lambda x: x[1], reverse=True)

sunny_rcm=[]
# 맑은 날과 판매된 상품의 상관 계수, p-value
for elem in result_sun: 
    pd_c = elem[0] 
    name = product_df[product_df["pd_c"] == int(pd_c)]
    name = name["clac_nm3"].values[0]
    print(elem, name)
    if elem[1] > 0.3 and elem[2] < 0.05:
        sunny_rcm.append(name)
```

```python
#구름많음, 판매된 상품 상관 관계 분석 (Point Biserial correlation)
cm_info = weather_df["weather_구름많음"]
result_dict_cm = dict()
for pd_c in pd_dict: 
    current = pd_dict[pd_c] 
    pb_cm = pointbiserialr(cm_info, current)
    result_dict_cm[pd_c] = pb_cm
    
result_cm = list() 
for key, val in result_dict_cm.items():
    if not math.isnan(val[0]):
        if val[1] <= 0.05: result_cm.append((key, *val))
result_sun.sort(key=lambda x: x[1], reverse=True)

cm_rcm=[]
# 구름많은 날과 판매된 상품의 상관 계수, p-value
for elem in result_cm: 
    pd_c = elem[0] 
    name = product_df[product_df["pd_c"] == int(pd_c)]
    name = name["clac_nm3"].values[0]
    print(elem, name)
    if elem[1] > 0.3 and elem[2] < 0.05:
        cm_rcm.append(name)
```
