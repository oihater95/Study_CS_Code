```python
import pandas as pd 
data = pd.read_csv("./data/weather_info.csv", encoding='cp949')
weather_df = pd.DataFrame(data) 
weather_df.head(20)
# 날씨 데이터 object형 -> binary로 변환 필요
```

```python
#날씨 데이터 binary변환
tmp_df = weather_df[["date", "weather"]] 
tmp_df = pd.get_dummies(tmp_df)
tmp_df.head()
```

```python
weather_df = pd.merge(left=weather_df, right=tmp_df, on="date") 
weather_df.pop("weather")
weather_df.head()
```

```python
#거래 정보 데이터
data2 = pd.read_csv("./data/data_transaction.csv", encoding='utf-8')
trans_df = pd.DataFrame(data2) 
trans_df.sort_values(by=['de_dt'], axis=0)
trans_df.rename(columns = {"de_dt" : "date"}, inplace = True)
trans_df.head(10)
```

```python
# 날씨 데이터와 거래정보 데이터 병합 (기준 : 날짜)
join_df = pd.merge(left=trans_df, right=weather_df, on="date", how="left")
join_df = join_df.sort_values(by=["date", "de_tm", "clnt_id"])
join_df = join_df[join_df["pd_c"] != "unknown"]
join_df.head(10)
```

```python
# 날짜 list & dictionary
date_len = len(join_df["date"].unique())
date_list = [elem for elem in join_df["date"].unique()]
date_list.sort()
date_dict = dict()
for idx, elem in enumerate(date_list): 
    date_dict[elem] = idx
date_len, date_list
```

```python
# 상품코드 dictionary & 날짜 dictionary 매칭
pd_dict = dict() 
for i in range(len(join_df)): 
    row = join_df.iloc[i] 
    pd_c, date = row["pd_c"], row["date"]
    if pd_c not in pd_dict: pd_dict[pd_c] = [0]*date_len
    pd_dict[pd_c][date_dict[date]] += 1
pd_dict
```

* 비와 상품 판매와의 관계

```python
#점이연 상관관계 point biserial (연속형 변수 : 이분형 변수 상관계수)
from scipy.stats import pointbiserialr

#비와 판매된 상품 상관 관계 분석
rain_info = weather_df["weather_비"]
result_dict = dict()
for pd_c in pd_dict: 
    current = pd_dict[pd_c] 
    pb = pointbiserialr(rain_info, current)
    result_dict[pd_c] = pb
```

```python
import math 
result = list() 
for key, val in result_dict.items():
    if not math.isnan(val[0]):
        if val[1] <= 0.05: result.append((key, *val))
result.sort(key=lambda x: x[1], reverse=True)
result
```

```python
# 상품코드 -> 상품 소분류명으로 변환
data = pd.read_csv("./data/data_product classification.csv", encoding='cp949')
product_df = pd.DataFrame(data) 
product_df = product_df[["pd_c", "clac_nm3"]]
product_df.head()
```

```python
# 비오는 날과 판매된 상품의 상관 계수, pvalue
for elem in result: 
    pd_c = elem[0] 
    name = product_df[product_df["pd_c"] == int(pd_c)]
    name = name["clac_nm3"].values[0]
    print(elem, name)
```

```python
# 미세먼지와 판매된 상품 상관관계 Pearson 사용 (연속형 변수 : 연속형 변수)
from scipy.stats import pearsonr
dust_info = weather_df["ave_dust"]
nan_idx = list() 
for idx, elem in enumerate(dust_info):
    if math.isnan(elem): 
        nan_idx.append(idx) 
dust_info = dust_info.dropna()
nan_idx, len(dust_info)
```

```python
result_dict = dict()
for pd_c in pd_dict: 
    current = pd_dict[pd_c] 
    current = current[:68] + current[70:]
    ps = pearsonr(dust_info, current)
    result_dict[pd_c] = ps
result_dict
result = list() 
for key, val in result_dict.items():
    if not math.isnan(val[0]):
        if val[1] <= 0.05: result.append((key, *val))
result.sort(key=lambda x: x[1], reverse=True)

for elem in result: 
    pd_c = elem[0] 
    name = product_df[product_df["pd_c"] == int(pd_c)]
    name = name["clac_nm3"].values[0]
    print(elem, name)
    #미세먼지 데이터는 판매된 상품과 상관관계가 있는 상품들이 적음을 알 수 있음
```
